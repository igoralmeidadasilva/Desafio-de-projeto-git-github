# **Desafio de projeto - Análise de dados com Python e Pandas**

**Análise exploratória na base de dados "Video game ratings from imdb"**

```python
# Importando as bibiotecas usadas no desafio
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importando o arquivo CSV
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/datasets/books_scraped.csv") 
df.head(10) # Visualização do inicio do Data frame
```

**Formatando o Dataframe**

Renomeando o cabeçalho para o português

```python
df = df.rename(columns={'Title': 'Título', 'Book_category': 'Categorias', 'Star_rating': 'Nota', 'Price': 'Preço', 'Stock': 'Estoque', 'Quantity': 'Quantidade'})
df.head(10)
```

**Verificando os tipos do Dataframe**

```python
df.dtypes
```

**Listando somente os livros com melhor avaliação**

```python
df.loc[df['Nota'] == 'Five']
```

**Organizando o dataframe por categoria dos livros**

```python
df.sort_values(by='Categoria')
```

**Contando a quantidade de livros por categoria**

Nota-se que, uma categoria indefinida, chamada "Add a comment" está presente.

```python
df1 = df['Categoria'].value_counts()
df1
```

**Excluindo uma linha com o tipo indefinido**

```python
df1 = df1.drop("Add a comment")
df1.head()
```

**Agora colocando no formato de um gráfico**

```python
#Figura
plt.figure(figsize=(25,5))
plt.title("Quantidade de livros por categoria")
plt.xlabel('Categoria')
plt.ylabel('Quantidade')

df1.plot.bar(color='darkblue');

```

**Comparação de preços**

Irei Comparar o preço dos livros de maior nota, porém como são muitos livros, irei usar uma amostra de apenas cinco

```python
# Separando a amostra
df_valor = df.loc[df['Nota'] == 'Five']
df_valor = df_valor.head()
```

```python
#Figura
plt.figure(figsize=(25,5))
plt.title("Compararação dos preços")
plt.plot(df_valor['Título'], df_valor['Preço']);
```
